```python
import gradio as gr
from transformers import GPT2Tokenizer, GPT2LMHeadModel
tokenizer = GPT2Tokenizer.from_pretrained("gpt2-medium")
model = GPT2LMHeadModel.from_pretrained("gpt2-medium")
model.config.pad_token_id = tokenizer.eos_token_id

def generate_text(prompt):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(
        **inputs,
        max_length=200,
        do_sample=True,
        top_k=50,
        top_p=0.95,
        temperature=0.7,
        repetition_penalty=1.2
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

interface = gr.Interface(
    fn=generate_text,
    inputs=gr.Textbox(label="Enter prompt", lines=2, placeholder="Type something..."),
    outputs=gr.Textbox(label="Generated text", lines=15, interactive=True, placeholder="GPT-2 output here"),
    title="GPT-2 Text Generator",
    description="Generate coherent text using GPT-2 Medium with reduced repetition"
)

interface.launch()
```
